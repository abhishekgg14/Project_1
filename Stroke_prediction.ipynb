```python
data=pd.read_csv('/content/healthcare-dataset-stroke-data.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
bmiMedianValue = data["bmi"].median()
print(bmiMedianValue)
```

```python
data["bmi"].fillna(bmiMedianValue,inplace=True)
data.isnull().sum()
```

```python
gender_dummies = pd.get_dummies(data["gender"])
married_dummies = pd.get_dummies(data["ever_married"])
work_dummies = pd.get_dummies(data["work_type"])
residence_dummies = pd.get_dummies(data["Residence_type"])
smoking_dummies = pd.get_dummies(data["smoking_status"])
data.columns
```

```python
stroke_df = data[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi','stroke']]
df = pd.concat([data, gender_dummies, married_dummies, work_dummies, residence_dummies, smoking_dummies], axis="columns")
df.columns
```

```python
df = df[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi',
       'stroke', 'Female', 'Male', 'No','Govt_job',
       'Never_worked', 'Private', 'Self-employed', 'Rural', 'formerly smoked', 'never smoked', 'smokes']]
df.head()
```

```python
y=df['stroke']
x=df.drop('stroke',axis=1)
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2)
```

```python
import tensorflow as tf
ann=tf.keras.Sequential()

ann.add(tf.keras.layers.Dense(units=25,activation='relu'))

ann.add(tf.keras.layers.Dense(units=25,activation='relu'))

ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

ann.compile('adam','binary_crossentropy',metrics=['accuracy'])
```

```python
result=ann.fit(x_train,y_train,epochs=10)
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

y_pred=[]
for i in ann.predict(x_test):
    if i>0.5:
        y_pred.append(1)
    if i<0.5:
        y_pred.append(0)
confusion_matrix(y_test,y_pred)
# Now, calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(conf_matrix)
```
